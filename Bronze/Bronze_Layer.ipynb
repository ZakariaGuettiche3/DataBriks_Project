```python
from pyspark.sql import functions as F
from delta.tables import DeltaTable
```

```python
dbutils.widgets.text("catalog","databriks","Catalog")
dbutils.widgets.text("data_source","brands","Data Source")

catalog = dbutils.widgets.get("catalog")
data_source = dbutils.widgets.get("data_source")

base_path = f"s3://zg-databriks/{data_source}/*.csv"
save_path = f"{catalog}.bronze.{data_source}"
```

```python
df =  spark.read.format("csv").option("header","true").option("inferSchema","true").load(base_path)\
    .withColumn("date_read",F.current_timestamp())\
    .withColumn("data_source",F.lit(data_source))\
    .select("*","_metadata.file_size")
    
display(df)
    
```

```python
df.write.format("delta").option("delta.enableChangeDataFeed","true").mode("overwrite")\
    .saveAsTable(save_path)
```
