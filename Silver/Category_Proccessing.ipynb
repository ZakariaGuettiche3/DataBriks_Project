```python
from pyspark.sql import functions as F
from delta.tables import DeltaTable
```

```python
df = spark.table("databriks.bronze.category")
display(df)
```

```python
df.printSchema()
```

```python
df = df.withColumn("category_code",F.upper(F.col("category_code"))).dropDuplicates(["category_code"])

display(df)
```

```python
df.write.format("delta")\
    .option("delta.enableChangeDataFeed", "true")\
    .mode("overwrite")\
    .saveAsTable("databriks.silver.category")
```
