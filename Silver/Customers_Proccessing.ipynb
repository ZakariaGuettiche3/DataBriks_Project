```python
from pyspark.sql import functions as F
from delta.tables import DeltaTable
```

```python
df = spark.table("databriks.bronze.customers")
display(df)
```

```python
df = df.withColumn("phone", F.when(F.col("phone").isNull(), F.lit("Not Available")).otherwise(F.col("phone").cast("string")))
```

```python

display(df.filter("phone = 'Not Available'"))
```

```python
df.filter(F.col("customer_id").isNull()).show(3)
```

```python
df = df.filter(F.col("customer_id").isNotNull())
```

```python
df.filter(F.col("customer_id").isNull()).show()
```

```python
df.write.format("delta")\
    .option("delta.enableChangeDataFeed", "true")\
    .mode("overwrite")\
    .saveAsTable("databriks.silver.customers")
```
