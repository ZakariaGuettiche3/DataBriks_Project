```python
from pyspark.sql import functions as F
from delta.tables import DeltaTable
```

```python
df = spark.table("databriks.bronze.brands")
display(df)
```

```python
df = df.withColumn("brand_name",F.trim(F.col("brand_name")))
display(df.select("brand_name").distinct())
```

```python
df = df.withColumn("brand_code", F.regexp_replace(F.col("brand_code"), "[^a-zA-Z]", ""))
display(df.select("brand_code").distinct())

```

```python
display(df.select("category_code").distinct())
```

```python
mapping = {
    "BOOKS": "BKS",
    "TOYS": "TOY",
    "GROCERY": "GRCY"
}

df = df.replace(mapping, "category_code")
display(df.select("category_code").distinct())
```

```python
display(df.limit(5))
```

```python
df.write.format("delta")\
    .option("delta.enableChangeDataFeed", "true")\
    .mode("overwrite")\
    .saveAsTable("databriks.silver.brands")
```
